Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points =np.array([4,7,2,10,13])
```

```python
plt.plot(points)
```

```python
a2=np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle='dashed')
plt.show()
```
